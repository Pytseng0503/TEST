```
import pandas as pd
import numpy as np
import requests
import csv
import re
```

```

def get_market_value():

    Data_final = pd.DataFrame([])
    for i in range(97,112):

      for j in ["01","02","03","04","05","06","07","08","09","10","11","12"]:

          for k in range(28,32):
              try :

                  url = "https://www.tpex.org.tw/web/stock/aftertrading/daily_mktval/mkt_result.php?l=zh-tw&d="+str(i)+"/"+str(j)+"/"+str(k)+"&s=0,asc,0&o=htm"
                  data = pd.read_html(url,header=None)

                  row = int(re.findall(r'\d+',data[0].iloc[-1][2])[0])
                  data = pd.DataFrame(data[0].iloc[0:row,0:6])
                  data.columns=data.columns.droplevel()
                  data['資料日期'] = str(i)+str(j)+str(k)
                  Data_final =pd.concat([Data_final,data],axis =0)


              except :
                    pass

    return Data_final
```

```
data = get_market_value()
```

```
data.to_excel('櫃買.xlsx')
```
